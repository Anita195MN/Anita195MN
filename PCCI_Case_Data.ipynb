```python
cat_cols = [ 'depression', 'sex',
       'ethnic_group_c', 'marital_status_c', 'insurance_provider', 'icu_yn',
       'tobacco_user', 'drugabuse',  'diabetes', 'anxiety',
       'obesity', 'dementia', 'persmentaldis',  'Condition',
       'care_plan_following_discharge',]
```

```python
num_cols = ['age', 'bmi', 'bp_diastolic',
       'bp_systolic', 'weight', 'height', 'pulse', 'temperature',
       'respirations', 'pat_pain_score_c', 'ed_visits', 'ip_visits',
       'chronic_conditions', 'meds_nutrition', 'glucose', 
       'potassium', 'sodium', 'artbloodgas', 'hemoglobin', 'chest_tube',
       'vent', 'los', 'LACE_Score',  'APRDRG',
       'cost_of_initial_stay',
       'care_plan_costs', 'cost_of_readmission']
```

```python
# Setting up the classification model
classification_model_setup = setup(df1, target = 'readmit30', fix_imbalance = True, categorical_features = cat_cols, numeric_features = num_cols)
```

```python
# Comparing the model
compare_models()
```

```python
# Creating amodel with the best algorithm
model = create_model('rf')
```

```python
# Evaluating the model
evaluate_model(model)
```

```python
# Loading the client daset for which prediction to be made
url1 = "https://raw.githubusercontent.com/Anita195MN/Anita195MN/main/PCCI%20Case%20Data%202.csv"
pred_data = pd.read_csv(url1)
```

```python
# Checking the data size of the client dataset
pred_data.shape
```

```python
# predicting the values using the best model
predictions = predict_model(model, data = pred_data)
```
