```python
# Code snippet 8
# Missing values
df.isna().sum()
```

```python
# Code snippet 9
# imputing missing values
mode = df1['CreditScore'].value_counts().index[0]
df1['CreditScore'].fillna(value = mode, inplace = True)
mode = df1['Geography'].value_counts().index[0]
df1['Geography'].fillna(value = mode, inplace = True )
mode = df1['Gender'].value_counts().index[0]
df1['Gender'].fillna(value = mode, inplace = True)
mode = df1['EstimatedSalary'].value_counts().index[0]
df1['EstimatedSalary'].fillna(value = mode, inplace = True )
```

```python
# Code snippet 10
df1['Balance'] = df1['Balance'].astype(str)
df1['EstimatedSalary'] = df1['EstimatedSalary'].astype(str)
```

```python
# code snippet 1
# importing pandas and numpy
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/Anita195MN/Anita195MN/main/Veridian%20Bank%20Customer%20Data.csv.crdownload'
```

```python
df = pd.read_csv(url)
```

```python
cols = ['CreditScore', 'Geography', 'Gender', 'Age',
       'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
       'EstimatedSalary', 'Exited']
```

```python
df1 = df[cols]
df1
```

```python
# code snippet 2
# importing all classification ML packages
from pycaret.classification import *
```

```python
# Code snippet 3
# set the categorical columns
cat_cols = ['Geography', 'Gender', 'HasCrCard', 'IsActiveMember',]
```

```python
# Code snippet 4
# Set the numerical columns
num_cols = ['CreditScore',  'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
```

```python
# Code snippet 5
# Set up
classification_model_setup = setup(df1, target = 'Exited', categorical_features = cat_cols, numeric_features = num_cols ) 
```

```python
# Code snippet 6
# Comparing model
compare_models(budget_time = 5)
```

```python
# Code snippet 7
# Creating a model using the best ML algorithm
model = create_model('gbc')
```

```python
# Code snippet 8
# Checking the confusion metrix etc
evaluate_model(model)
```
